# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
l_regressor = LinearRegression()
l_regressor.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
p_regressor = PolynomialFeatures(degree=2)
X_poly = p_regressor.fit_transform(X)

l_regressor2 = LinearRegression()
l_regressor2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```
plt.scatter(X, y, color='red')
plt.plot(X, l_regressor.predict(X), color='blue')
plt.title('Actual Salary vs Predicted Salary (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, y, color='red')
plt.plot(X, l_regressor2.predict(X_poly), color='blue')
plt.title('Actual Salary vs Predicted Salary (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

### Creating new model to train with a different 'n' value (different degrees).

```
#Creating new Polynomial features with different degree.
p_regressor2 = PolynomialFeatures(degree=4)
X_poly2 = p_regressor2.fit_transform(X)

#Creating new linear regression equation we will pass in the new X values.
l_regressor3 = LinearRegression()
l_regressor3.fit(X_poly2, y)

#Plotting these new results to compare.
plt.scatter(X, y, color='red')
plt.plot(X, l_regressor3.predict(X_poly2), color='blue')
plt.title('Actual Salary vs Predicted Salary (Polynomial Regression Retrained)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
#Only done for course purposes. Make visualization look better.
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, l_regressor3.predict(p_regressor2.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

```
l_regressor.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
l_regressor3.predict(p_regressor2.fit_transform([[6.5]]))
```
